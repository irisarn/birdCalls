```python
url = "http://www.xeno-canto.org/api/2/recordings?"
country = '"United States"'
state = "Oregon"
queryURL = f"{url}query=cnt:{country}+loc:{state}"
response = requests.get(queryURL).json()
```

```python
print(json.dumps(response, indent=4, sort_keys=True))
#urllib.request.urlretrieve
```

```python
#Empty lists to append information to
ID = []
Name = []
Genus = []
Location = []
lat = []
lon = []
Quality = []
Remarks = []
recordingType = []
downloadURL = []

#Loop through each of the results in response to pull the date wanted
for bird in range(len(response["recordings"])):
    try:
      response = requests.get(queryURL).json()
      ID.append(response['recordings'][bird]['id'])
      Name.append(response['recordings'][bird]['en'])
      Genus.append(response['recordings'][bird]['gen'])
      Location.append(response['recordings'][bird]['loc'])
      lat.append(response['recordings'][bird]['lat'])
      lon.append(response['recordings'][bird]['lng'])
      Quality.append(response['recordings'][bird]['q'])
      Remarks.append(response['recordings'][bird]['rmk'])       
      recordingType.append(response['recordings'][bird]['type'])
      downloadURL.append(response['recordings'][bird]['file'])
    except:
        print(f"Processing Record {bird} of Set 1 | {response['recordings'][bird]['en']}") 
```

```python
#Store data into dataframe
birdData = pd.DataFrame(ID)
birdData.rename(columns={0:"id"}, inplace=True)
birdData["English Name"] = Name
birdData["Genus"] = Genus
birdData["Location"] = Location
birdData["Latitude"] = lat
birdData["Longitude"] = lon
birdData["Quality Score"] = Quality
birdData["Author Remarks"] = Remarks
birdData["Recording Type"] = recordingType
birdData["MP3 Download URL"] = downloadURL
birdData.head(10)
```

```python
#Export to csv file
birdData.to_csv("birdData.csv")
```
